```python
import re #regular expression
import pandas as pd
```

```python
f = open('chat_24h_no_seconds.txt','r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
#regex101.com
##pattern = '\[\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\]\s'
pattern = r'\[\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\]\s'
```

```python
messages = re.split(pattern,data)[1:]
#messages
```

```python
dates = re.findall(pattern,data)
#dates
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})
#convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'],format='[%d/%m/%y, %H:%M] ')
df.rename(columns = {'message_date':'date'},inplace = True)
#df.head()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

#df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df['month'] = df['date'].dt.month_name()
```

```python
df['day'] = df['date'].dt.day
```

```python
df['hour'] = df['date'].dt.hour
```

```python
df['minute'] = df['date'].dt.minute
```

```python
from collections import Counter
f = open('stop_hinglish.txt','r')
stop_words = f.read()

temp = df[df['user'] != 'F.R.I.E.N.D.S IITH']
# Remove all messages containing 'omitted'
temp = temp[~temp['message'].str.contains(
    'omitted',
    case=False,
    na=False
)]

words = []

for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)

most_common_df = pd.DataFrame(Counter(words).most_common(20))
print(Counter(words).most_common(10))
```
